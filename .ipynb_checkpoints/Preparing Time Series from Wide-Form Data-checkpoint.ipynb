```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mticks
import seaborn as sns
import datetime
from matplotlib.ticker import FuncFormatter

import missingno as miss
import datetime as dt
import statsmodels.tsa.api as tsa
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv')
```

```python
df.head()
```

# Part 1

```python
cities_df = df[df['SizeRank'].between(0,3, inclusive='both')]
cities_df.head()
```

```python
cities_df.info()
```

```python
id_vars = cities_df.columns[:8]
id_vars
```

```python
cities_long = pd.melt(cities_df, id_vars=id_vars, var_name='Date')
cities_long
```

```python
cities_long['Date'][2]
```

```python
fmt = '%Y-%m-%d'
cities_long['Date'] = pd.to_datetime(cities_long['Date'], format=fmt)
cities_long.info()
```

```python
cities_long = cities_long.set_index('Date')
cities_long
```

```python
cities_ts = cities_long.groupby('RegionName').resample("M").last()['value']
cities_ts
```

# Part 2

```python
cities_unstacked = cities_long.copy()
cities_unstacked = cities_ts.unstack(0)
cities_unstacked
```

```python
cities_unstacked.plot();
```

```python
fig, ax = plt.subplots()

ax.plot(cities_unstacked)

def thousands(x,pos):
    return f'{x/1000:.0f}K'

ax.yaxis.set_major_formatter(FuncFormatter(thousands))

ax.set_xlabel("Year")
ax.set_ylabel('Value')

ax.legend(cities_unstacked.columns[0:])

plt.show()
```
